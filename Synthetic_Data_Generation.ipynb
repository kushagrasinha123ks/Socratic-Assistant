```python
from huggingface_hub import HfApi, HfFolder
```

```python
df_socratic50k = pd.read_csv('/content/socratic_data50k.csv')
```

```python
df_socratic50k.info()
```

```python
df_socratic50k.dropna(inplace=True)
df_socratic50k.info()
```

```python
dataset = Dataset.from_pandas(df_socratic50k)

def transform_conversation(example):
    user_text = example['user']
    assistant_text = example['assistant']

    return {'text': f'<s>[INST] {user_text} [/INST] {assistant_text} </s>'}
```

```python
transformed_dataset = dataset.map(transform_conversation)
```

```python
final_dataset = DatasetDict({
    'train': transformed_dataset,
})
```

```python
print(final_dataset['train'].column_names)
```

```python
final_dataset = final_dataset['train'].remove_columns([col for col in final_dataset['train'].column_names if col != 'text'])
```

```python
print(len(final_dataset))
```

```python
hf_token = HfFolder.get_token()
api = HfApi()

dataset_name = 'kush19/socratic_dataset50k'
api.create_repo(dataset_name, repo_type="dataset")

final_dataset.push_to_hub(dataset_name, token=hf_token)

print(f'Dataset uploaded to https://huggingface.co/{dataset_name}')
```
